# Questão 2:

Mostrar que $$\frac{e^{1/x}}{1+e^{1/x}} = \frac{1}{e^{-1/x}+1}$$

Podemos multiplicar a primeira equação por $$\frac{e^{1/x}}{e^{1/x}}$$ 
Assim: $$\frac{e^{1/x}}{1+e^{1/x}} * \frac{e^{1/x}}{e^{1/x}} = \frac{\frac{e^{1/x}}{e^{1/x}}}{\frac{1 + e^{1/x}}{e^{1/x}}} =  \frac{1}{\frac{1}{e^{1/x}} + 1} = \frac{1}{e^{-1/x}+1}$$

Agora, para testar qual das duas equações funciona para x muito pequenos, temos:


```python
import math

print ("Pela primeira equação:\n")

for i in range (100,0,-1):
    x = 0.001*i
    print ("x = %.20f" % x, ": ", math.exp(1.0/x)/(1 + math.exp(1.0/x)))
    
    
```

Percebemos que para a primeira equação, ocorreu um erro de Overflow para x>0.002.

Agora, pela segunda equação:

```python
import math

print ("Pela segunda equação:\n")

for i in range (100,0,-1):
    x = 0.001*i
    print ("x = %.20f" % x, ": ", (1)/(math.exp(-1.0/x)+1))
```

Já a segunda equação não apresentou erros com x entre 0.1 e 0.001. 
Portanto a equação $$\frac{1}{e^{-1/x}+1}$$ é mais indicada para valores de x pequenos.
